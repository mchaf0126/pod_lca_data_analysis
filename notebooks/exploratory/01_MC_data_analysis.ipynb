```python
from pathlib import Path
import pandas as pd
```

```python
main_directory = Path(__vsc_ipynb_file__).parents[2]
raw_bom_path = main_directory.joinpath('data/raw/raw_boms.xlsx')
all_str_bom = pd.read_excel(raw_bom_path, sheet_name='Structure')
all_enc_o_bom = pd.read_excel(raw_bom_path, sheet_name='Enclosure - Opaque')
all_enc_t_bom = pd.read_excel(raw_bom_path, sheet_name='Enclosure - Translucent')
all_enc_r_bom = pd.read_excel(raw_bom_path, sheet_name='Enclosure - Roofing')
```

```python
for str_option in all_str_bom['Option'].unique():
    for enc_o_option in all_enc_o_bom['Option'].unique():
        for enc_t_option in all_enc_t_bom['Option'].unique():
            new_folder = main_directory.joinpath(f'data/template_models/{str_option}_{enc_o_option}_{enc_t_option}_ENCR1')
            new_folder.mkdir(parents=True, exist_ok=True)
            bom_folder = new_folder.joinpath('bom')
            bom_folder.mkdir(parents=True, exist_ok=True)
            impacts_folder = new_folder.joinpath('impacts')
            impacts_folder.mkdir(parents=True, exist_ok=True)
            prebuilt_scenarios_folder = new_folder.joinpath('prebuilt_scenarios')
            prebuilt_scenarios_folder.mkdir(parents=True, exist_ok=True)
            
```

```python
for str_option in all_str_bom['Option'].unique():
    current_str_option = all_str_bom[all_str_bom['Option'] == str_option].copy()
    for enc_o_option in all_enc_o_bom['Option'].unique():
        current_enc_o_option = all_enc_o_bom[all_enc_o_bom['Option'] == enc_o_option].copy()
        for enc_t_option in all_enc_t_bom['Option'].unique():
            current_enc_t_option = all_enc_t_bom[all_enc_t_bom['Option'] == enc_t_option].copy()
            temp_df = pd.concat(
                [
                    current_str_option,
                    current_enc_o_option,
                    current_enc_t_option,
                    all_enc_r_bom
                ]
            )
            temp_df.to_csv(
                main_directory.joinpath(f'data/template_models/{str_option}_{enc_o_option}_{enc_t_option}_ENCR1/bom/{str_option}_{enc_o_option}_{enc_t_option}_ENCR1_bom.csv'),
                index=False
            )
            
            
```

```python
for str_option in all_str_bom['Option'].unique():
    for enc_o_option in all_enc_o_bom['Option'].unique():
        for enc_t_option in all_enc_t_bom['Option'].unique():
            print(f'{str_option}_{enc_o_option}_{enc_t_option}_ENCR1')
```
